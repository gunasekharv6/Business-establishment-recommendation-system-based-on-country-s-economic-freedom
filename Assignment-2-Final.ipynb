<b>Using REGEX, populating lists with the extracted data from tweets</b><br>
<b> These lists will be used to populate database tables further</b>

```python
database.commit()
```

```python
economicRanking = []
dominantIndustry = []
labourCost = []
tags=[]

for tweet in user_tweets:
    print(tweet.full_text)
    print("\n")
    for country in countries.values():
        if country.lower() in tweet.full_text.lower():
            tgs=re.findall("#.* ", tweet.full_text);
            if(len(tgs)!=0):
                for tag in tgs:
                    tags.append((tweet.id, tag))
            
            y = re.findall("by the .* sector", tweet.full_text); 
            if(len(y)!=0):
                dominantIndustry.append((country, y[0][7:-7]))
             
            
            z = re.findall("by the .* industry", tweet.full_text);
            if(len(z)!=0):
                dominantIndustry.append((country, z[0][7:-8]))
            
            if("average labor costs" in tweet.full_text):
                x2 = re.findall("\d+\.?\d*", tweet.full_text);
                if(len(x2)!=0):
                    labourCost.append((country, x2[0], x2[1]))
            else:
                x = re.findall("\d+\.?\d*", tweet.full_text);#"\d{1,3}\.?\d*"
                if(len(x)!=0):
                    economicRanking.append((country, x[1], x[2]));
              
            
            
```
